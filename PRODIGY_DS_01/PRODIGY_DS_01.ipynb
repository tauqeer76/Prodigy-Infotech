```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\User\\Desktop\\New folder\\population.csv")
df
```

```python
df.head()
```

```python
df.describe
```

```python
countries_to_plot = ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France']
filtered_data = df[df['Country Name'].isin(countries_to_plot)]
plt.bar (filtered_data['Country Name'], filtered_data['2023'])
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2023')
plt.title('Population of selected countries in 2023')
plt.xticks(rotation=45, ha='right') 
plt.show()
```

```python
plt.hist(filtered_data['2023'], bins=10)
plt.xlabel('Population in 2023')
plt.ylabel('Frequency')
plt.title('Distribution of Population in 2023')
plt.xticks(rotation=45, ha='right')
plt.show()
```
